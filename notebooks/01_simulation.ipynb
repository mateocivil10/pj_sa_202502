```python
import random
import matplotlib.pyplot as plt

def has_duplicate_birthdays(num_people):
    """Simula los cumplea√±os de num_people personas y verifica si hay duplicados."""
    birthdays = [random.randint(1, 365) for _ in range(num_people)]
    return len(birthdays) != len(set(birthdays))

def monte_carlo_birthday_paradox(num_people, num_simulations):
    """Simula la paradoja del cumplea√±os usando Monte Carlo y devuelve la probabilidad de cumplea√±os compartidos."""
    duplicate_count = 0
    
    for _ in range(num_simulations):
        if has_duplicate_birthdays(num_people):
            duplicate_count += 1
    
    # Calcula la probabilidad de que al menos dos personas compartan el mismo cumplea√±os
    probability = duplicate_count / num_simulations
    return probability, duplicate_count, num_simulations - duplicate_count

# Par√°metros
num_people = 52  # N√∫mero de personas en la habitaci√≥n
num_simulations = 10000  # N√∫mero de simulaciones a realizar

# Ejecuta la simulaci√≥n de Montecarlo
probability, duplicate_count, non_duplicate_count = monte_carlo_birthday_paradox(num_people, num_simulations)

# Muestra el resultado
print(f"En un grupo de {num_people} personas, la probabilidad de que dos compartan cumplea√±os es aproximadamente {probability:.2%}")

# Datos para el gr√°fico de barras
casos = ['Success', 'Failure']
cantidad = [duplicate_count, non_duplicate_count]

# Crear el gr√°fico de barras
plt.figure(figsize=(10, 6))
plt.bar(casos, cantidad, color=['green', 'red'])

# A√±adir t√≠tulos y etiquetas
plt.title('% of success /Failure')
plt.xlabel('Case')
plt.ylabel('Amount')

# A√±adir etiquetas encima de las barras con porcentajes
for i, v in enumerate(cantidad):
    percentage = (v / num_simulations) * 100
    plt.text(i, v + 0.5, f'{percentage:.2f}%', ha='center', va='bottom')

# Mostrar el gr√°fico
plt.show()
```

```python
num_simulations = 10000  # N√∫mero de simulaciones a realizar
max_people = 70  # N√∫mero m√°ximo de personas en la habitaci√≥n

# Almacenar probabilidades para diferentes n√∫meros de personas
probabilities = []

# Ejecuta la simulaci√≥n de Montecarlo para diferentes n√∫meros de personas
for num_people in range(1, max_people + 1):
    probability = monte_carlo_birthday_paradox(num_people, num_simulations)
    probabilities.append(probability[0])

# Crear el gr√°fico de la curva
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_people + 1), probabilities, marker='o', linestyle='-', color='b')

# A√±adir t√≠tulos y etiquetas
plt.title('Probabilidad de Cumplea√±os Compartido')
plt.xlabel('N√∫mero de personas en la habitaci√≥n')
plt.ylabel('Probabilidad')

# Mostrar el gr√°fico
plt.grid(True)
plt.show()
```
